# Lets try to represent the image in grayscale using nth truncated SVD term

```python
def myCompressor(image, n):
    U,s,V = np.linalg.svd(image, full_matrices=False)
    SimpForm = np.dot(U[:,:n],np.dot(np.diag(s[:n]),V[:n,:]))
    return SimpForm
```

# Using $n=5$
You can almost make out Thea's head in the center, but the image is still quite hard to make out.

```python
A5 = myCompressor(arrgrey[:,:,0],5)
I5=Image.fromarray(A5)
fig = plt.figure(figsize=(10,8))
imshow(I5);
```

# How much does this help for storage?

```python
def howMany(i,m,n):
    l = i*(m+n+1) #Each term of series is uxvxs = mxnxs so we need m+n+1 values
    N = m*n #Total size of image array
    p=100*l/N
    print("Using",i,"terms of the series, we need",l,"values stored. Compare this to,",N,"values for the original matrix, or %.4f" % p,"% of the original amount")
```

```python
howMany(50,m,n)
print(m*n)
```

# Now, lets try the same thing with the RBG form of the matrix
This time, we do the compression on each of the R/G/B matrices, and then make a new image array with each slice being the corresponding color reconstruction

## We see that with only one term of the series, the image is almost completely indiscernable. Only the relative color gradient is visible

```python
r = myCompressor(arr[:,:,0],1)
b = myCompressor(arr[:,:,1],1)
g = myCompressor(arr[:,:,2],1)
A = np.empty([n,m,p], dtype=np.uint8)
A[:,:,0] = r
A[:,:,1] = b
A[:,:,2] = g
I=Image.fromarray(A)
fig = plt.figure(figsize=(10,8))
imshow(I);
```
