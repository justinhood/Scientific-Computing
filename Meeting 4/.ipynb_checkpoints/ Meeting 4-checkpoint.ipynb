```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# \# 3. DIY Matrix Multiplication

```python
def mymult(A,B):
    # Get the shapes of the matrices and prove that j=m for multiplication purposes
    i=np.shape(A)[0]
    j=np.shape(A)[1]
    m=np.shape(B)[0]
    n=np.shape(B)[1]
    C=np.empty([i,n])
    if(j!=m):
        return "Failure"
    for a in range(0,i):
        for b in range(0,n):
            C[a,b] = np.dot(A[a,:],B[:,b])
    return C
```

```python
A = np.array([[1,2],[3,4],[5,6]])
B = np.array([[7,9,11],[8,10,12]])
print("A=")
print(A)
print("B=")
print(B)
C=mymult(A,B)
print("C=")
print(C)
```

## Let,
A=\begin{bmatrix}
1 & 2 \\ 0 & 2
\end{bmatrix}
From 3.7

```python
A= np.array([[1,2],[0,2]])
l=4
# Construct a unit circle representing the terminal points of a set of unit vectors
npts = 200
x = np.linspace(-1,1,npts)
y = np.sqrt(1-x**2)
# Construct unit vectors that trace out the unit circle
circ = np.hstack([np.array([x, y]),np.array([-x, -y])]) #<-- stack to top and bottom of the circle
# Construct the ellipse by multiplying A by the unit (circle) vectors
b = np.dot(A,circ)
# Display the results

v,P = np.linalg.eig(mymult(np.transpose(A),A))
idx = v.argsort()[::-1]
v=v[idx]
P=P[:,idx]
Sig=np.zeros([np.shape(A)[1],np.shape(v)[0]])
for i in range(0,np.shape(v)[0]):
    Sig[i,i]=np.sqrt(v[i])
V=mymult(Sig,P)

U=mymult(A,np.linalg.inv(mymult(Sig,P)))

mx = np.argmax(abs(v))
mn = np.argmin(abs(v))

# Axes created by eigenvectors -- scaled by eigenvalues
# Construct the minor and major axes of the ellipse b = A*xx (from above)
minorx = np.linspace(0,np.sqrt(v[mn])*U[0,mn],npts)
minory = np.linspace(0,np.sqrt(v[mn])*U[1,mn],npts)
majorx = np.linspace(0,np.sqrt(v[mx])*U[0,mx],npts)
majory = np.linspace(0,np.sqrt(v[mx])*U[1,mx],npts)


circminorx = np.linspace(0,P[0, mn],npts)
circminory = np.linspace(0,P[1, mn],npts)
circmajorx = np.linspace(0,P[0, mx],npts)
circmajory = np.linspace(0,P[1, mx],npts)

plt.plot(circ[0,:],circ[1,:],'b')
plt.axis([-2,2,-2,2]);
plt.legend(['unit circle $x$'],loc='lower right');
plt.plot(circminorx, circminory, 'r', circmajorx, circmajory, 'r');
plt.savefig('37circ.png')

plt.plot(circ[0,:],circ[1,:],'b',b[0,:],b[1,:],'r')
plt.axis([-l,l,-l,l]);
plt.legend(['unit circle $x$','$b = Ax$'],loc='lower right');
plt.plot(minorx,minory,'k', majorx,majory,'k',circminorx, circminory, 'r', circmajorx, circmajory, 'r');
plt.savefig('37elip.png')
```
